```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
games=pd.read_csv(r"C:\Users\umesh\OneDrive\Desktop\games.csv")
games
```

```python
sales=pd.read_csv(r"C:\Users\umesh\OneDrive\Desktop\games sales.csv")
sales
```

```python
data=(games,sales)
```

```python
vg1=pd.concat(data)
vg1
```

```python
vg1.describe(include="all").T
```

```python
vg1.info()
```

```python
vg1.isnull().sum()
```

```python
vg2=vg1.copy()
vg2
```

```python
vg2.describe(include='all')
```

```python
vg2['NA_Sales'].fillna(vg2['NA_Sales'].mean(), inplace=True)
vg2['EU_Sales'].fillna(vg2['EU_Sales'].mean(), inplace=True)
vg2['JP_Sales'].fillna(vg2['JP_Sales'].mean(), inplace=True)
vg2['Other_Sales'].fillna(vg2['Other_Sales'].mean(), inplace=True)
vg2['Global_Sales'].fillna(vg2['Global_Sales'].mean(), inplace=True)
```

```python
vg2.describe(include='all')
```

```python
vg2.isnull().sum()
```

```python
df=vg2.dropna()
df
```

```python
df.describe(include='all').T
```

```python
df.isnull().sum()
```

```python
# create a new column 'Total_Sales' as the sum of 'NA_Sales', 'EU_Sales', and 'JP_Sales'
df['Total_Sales'] = (df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'])
df['Total_Sales']
```

```python
df
```

```python
df["Year"] = df["Year"].astype(int)
print(df.dtypes)
```

```python
vg=df.copy()
```

```python
vg.info()
```

```python
vg.describe(include="all").T
```

```python
vg.columns
```

```python
genre_counts = vg['Genre'].value_counts()
top_5_genres = genre_counts.head(5).index.tolist()
```

```python
filtered_data = vg[vg['Genre'].isin(top_5_genres)]
```

```python
sns.stripplot(x='Genre', y='Global_Sales', data=filtered_data, jitter=True)
plt.xlabel('Genre')
plt.ylabel('Global_Sales')
plt.title('Strip Plot of Genre vs. Global_Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
publisher_counts = vg['Publisher'].value_counts()
top_5_publishers = publisher_counts.head(5).index.tolist()
```

```python
filtered_data = vg[vg['Publisher'].isin(top_5_publishers)]
```

```python
sns.barplot(x='Publisher', y='Global_Sales', data=filtered_data, ci=None)
plt.xlabel('Publisher')
plt.ylabel('Global_Sales')
plt.title('Bar Plot of Publisher vs. Global_Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
Sales = vg.groupby(['Name'],as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales' , ascending=False).head(10)
sns.barplot(data=Sales,x='Name',y='Total_Sales') 
sns.set(rc={'figure.figsize':(30,8)})
```

```python
Sales = vg.groupby(['Genre'],as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales' , ascending=False).head(10)
sns.barplot(data=Sales,x='Genre',y='Total_Sales') 
sns.set(rc={'figure.figsize':(10,8)})
```

## insight: Video game genre plays a crucial role in sales. Certain genres may be more popular in specific regions or among particular age groups.

```python
Sales = vg.groupby(['Publisher'],as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales' , ascending=False).head(10)
sns.barplot(data=Sales,x='Publisher',y='Total_Sales') 
sns.set(rc={'figure.figsize':(25,10)})
```

## From above data, it indicates that the publisher has a strong reputation in the gaming industry.
## This reputation might be built on the basis of producing high-quality games with engaging content.

```python
Sales = vg.groupby(['Platform'],as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales' , ascending=False).head(10)
sns.barplot(data=Sales,x='Platform',y='Total_Sales') 
sns.set(rc={'figure.figsize':(10,5)})
```

## Here we see that the people are more likely play the games on DS and PS2.

```python
Sales = vg.groupby(['Year'],as_index=False)['Total_Sales'].sum().sort_values(by='Total_Sales')
sns.lineplot(data=Sales,x='Year',y='Total_Sales') 
```
